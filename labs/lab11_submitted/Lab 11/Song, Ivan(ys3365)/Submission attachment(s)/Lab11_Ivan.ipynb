```python
number_of_nests = [14522, 3261, 4448, 4517, 4633, 6929, 4879, 3429,2383,3321,\
                   2143,3790,3597,2626,2674,2107,2077,2055,1720,1596]
number_of_females = [2640,593,809,821,842,1260,887,623,433,604,390,689,654,477,\
                     486,383,378,374,313,290]
years = list(range(1984,2004))

# Question 1: Calculate What is the approximate number of nests in which a single female lays eggs.
print('Question 1 \n')
nest_of_one_female = [0] * len(number_of_females)
print("Approximate number of nests in which a single female lays eggs")
for index in range (len(nest_of_one_female)):
  nest_of_one_female[index] = number_of_nests[index] / number_of_females[index]
  print(years[index],format(nest_of_one_female[index], '.6f'), sep = ": ")
  
# Question 2: What is the percentage by which the number of nests differs in each year from the previous year?
print('\nQuestion 2 \n')
ratio = [0] * len(number_of_nests)
print("The percentage by which the number of nests differs in each year from the previous year")
for index in range (len(number_of_nests)):
  if index > 0:
    ratio[index] = number_of_nests[index] / number_of_nests[index - 1] * 100 - 100
  print(years[index], format(ratio[index], '.2f'), sep = ": ", end = "%\n")
average = sum(ratio) / len(ratio) - 1
print("Average: ", average)

# Question 3: Plot the number of nests and number of females as a function of years.
print('\nQuestion 3 \n')
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4), dpi=100, facecolor='gray')
plt.plot(years, number_of_nests, 'b')
plt.plot(years, number_of_females, 'r')
plt.plot(years, number_of_nests, 'g.')
plt.plot(years, number_of_females, 'g.')
plt.xlabel("years")
plt.ylabel("number of nests (blue) and females (red)")
plt.show()
```

## Problem 2: Fobonacci Numbers and the Golden Ratio (35 points) 


You just heard two different talks on the topic of Fibonacci numbers and the golden ratio. 

Write a program that creates a list of the first 50 Fibonacci numbers.

Once the numbers are calculated the program should calculate the ratio of the two consecutive numbers for each pair of Fibonacci numbers:

$$\dfrac{\text{Fibonacci number at position i}}{\text{Fibonacci number at position i-1}}$$

--

The __golden ratio__ can be calculated using the following formula:

$$\dfrac{1 + \sqrt{5}}{2} $$

Compare the value of the golden ratio to the values obtained by calculating the ratios of consecutive Fibonacci numbers. 



Create a CODE cell below to write your program. Your answers should be in a TEXT cell or in the comments in the program. 

```python
fib = [1] * 50
ratio = [0] * 50
golden_ratio = (1 + 5 ** (1/2))/2

print("FiboNacci Number", "Fibbonaci number ratio", "Golden Ratio", sep = "\t"*2)
print('-' * 100)

for index in range (len(fib)):
  if index > 1:
    fib[index] = fib[index - 1] + fib[index - 2]
  if index > 0:
    ratio[index] = format(fib[index] / fib[index - 1], '.6f')
  print(fib[index], ratio [index], golden_ratio, sep = "\t" * 4)
```

ANSWER: I first created a list for fibonacci numbers containing 50 "1"s, and a list for the list of values of ratio of the adjacent numbers containing 50 "0"s (Since first two numbers in fibonacci sequence are 1, and the first ratio should no exist, and shouldn't be represented by 1). Then, I decided to calculate the fibonacci numbers and ratios using a for loop: "for index in range (len(fib))" (which is basically in range (50)), and left the first two items in the fibonacci lists remain as "1" and the first item in the ratio list remain as "0" using two if statements; for all later items I used functions below:

fib[index] = fib[index - 1] + fib[index - 2]

ratio[index] = fib[index] / fib[index - 1]

Then I set a variable for golden ratio, and compared the list of ratios and the golden ratio in a table.

The result is that the bigger the fibonacci numbers get, the ratio of those adjacent numbers becomes closer to the golden ratio of 1.618.

There's also another trend. all the odd ratios (ratios in odd position of the list) are smaller than the actual golden ratio, and all even ratios are bigger than the golden ratio. The difference of both odd and even ratios to the golden ratio is decreasing when the fibonacci numbers get bigger.
