```python
# include the library for reading data from url
import urllib.request

# where should we obtain data from?
url_war_and_peace = "http://www.gutenberg.org/files/2600/2600-0.txt"
url_anna_karenina = "http://www.gutenberg.org/files/1399/1399-0.txt"
url_hamlet = "http://www.gutenberg.org/files/1524/1524-0.txt" 
url_mackbeth = "http://www.gutenberg.org/cache/epub/2264/pg2264.txt" 
url_pride_and_prejudice = "http://www.gutenberg.org/files/1342/1342-0.txt" 
url_sense_and_sensibility = "http://www.gutenberg.org/cache/epub/161/pg161.txt" 

# initiate request to URL
response_wp = urllib.request.urlopen(url_war_and_peace)
response_ak = urllib.request.urlopen(url_anna_karenina)
response_h = urllib.request.urlopen(url_hamlet)
response_m = urllib.request.urlopen(url_mackbeth)
response_pp = urllib.request.urlopen(url_pride_and_prejudice)
response_ss = urllib.request.urlopen(url_sense_and_sensibility)

# read data from URL as a string, making sure
# that the string is formatted as a series of ASCII characters
book_wp = response_wp.read().decode('utf-8')
book_ak = response_ak.read().decode('utf-8')
book_h = response_h.read().decode('utf-8')
book_m = response_m.read().decode('utf-8')
book_pp = response_pp.read().decode('utf-8')
book_ss = response_ss.read().decode('utf-8')


# split the book into individual words
words_wp = book_wp.split() 
words_ak = book_ak.split()
words_h = book_h.split()
words_m = book_m.split()
words_pp = book_pp.split()
words_ss = book_ss.split()



word_lengths_wp = [0]*30

for word in words_wp : 
    if len(word) < 30:
        word_lengths_wp[ len(word) ] += 1 
    

word_lengths_ak = [0]*30

for word in words_ak : 
    if len(word) < 30:
        word_lengths_ak[ len(word) ] += 1 
    

word_lengths_h = [0]*30

for word in words_h : 
    if len(word) < 30:
        word_lengths_h[ len(word) ] += 1 
    
    

word_lengths_m = [0]*30

for word in words_m: 
    if len(word) < 30:
        word_lengths_m[ len(word) ] += 1 
    
    

word_lengths_pp = [0]*30

for word in words_pp: 
    if len(word) < 30:
        word_lengths_pp[ len(word) ] += 1 
    
    

word_lengths_ss = [0]*30

for word in words_ss: 
    if len(word) < 30:
        word_lengths_ss[ len(word) ] += 1 
    


for i in range(30):
    word_lengths_wp[i] = word_lengths_wp[i]/ len(words_wp)
    word_lengths_ak[i] = word_lengths_ak[i]/ len(words_ak)
    word_lengths_h[i] = word_lengths_h[i]/ len(words_h)
    word_lengths_m[i] = word_lengths_m[i]/ len(words_m)
    word_lengths_pp[i] = word_lengths_pp[i]/ len(words_pp)
    word_lengths_ss[i] = word_lengths_ss[i]/ len(words_ss)


#####################3

import numpy as np
import matplotlib.pyplot as plt


#finally, plot the graph and show it 
plt.figure(figsize=(9,6), dpi=150)
plt.plot(word_lengths_wp, 'g-.', label="War and Peace")   
plt.plot(word_lengths_ak, 'b-.', label="Anna Karenina")   
plt.plot(word_lengths_h, 'r-.', label="Hamlet") 
plt.plot(word_lengths_m, 'y-.', label="Mackbeth")
plt.plot(word_lengths_pp, 'm-.', label="Pride and Prejudice")
plt.plot(word_lengths_ss, 'k-.', label="Sense and Sensibility")
plt.xlabel('word length')        # add title to the x-axis
plt.ylabel('count')    # add title to the y-axis 
plt.legend(loc='upper right')   # create the legend with plot descriptions 

plt.show()
```
